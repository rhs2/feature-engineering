## SMOTE(Synthetic Minority Oversampling Technique)
SMOTE (Synthetic Minority Over-sampling Technique) is a technique used in machine learning to address imbalanced datasets where the minority class has significantly fewer instances than the majority class. SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances.

- What Smote does?
-- in the upsampling data points are created on the minority datapoints existing locations, thats means it just increasing the data point not the variance. On the same point location it created more and more data. That is not also good, 

that's why we use SMOTE(synthetic minority Oversampling Technique) because what smote does is in the minority data points it makes a cross bridge line in 2 data points and in that line it creates synthetic data points with variance. and it just connects the existing datapoints and create more datapoints on the line bridge. 

```python
# to creating the dataset
from sklearn.datasets import make_classification
```

```python
X, y= make_classification(n_samples=1000, n_redundant=0, n_features=2, n_clusters_per_class=1,
                    weights=[0.90], random_state=12)
```

```python
X.shape
```

```python
y.shape
```

```python
import pandas as pd
df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
## visualizing the plot
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'], final_df['f2'], c= final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
oversample= SMOTE()
X,y = oversample.fit_resample(final_df[['f1', 'f2']], final_df['target'])
```
